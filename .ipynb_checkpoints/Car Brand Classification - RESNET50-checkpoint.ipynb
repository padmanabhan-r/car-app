### Importing all genaral libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob

%matplotlib inline
```

### Importing all TF libraries

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input, Lambda, Flatten
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
```

```python
import tensorflow as tf
#print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
tf.config.experimental.list_physical_devices('GPU')
tf.debugging.set_log_device_placement(True)
```

### Resnet

```python
IMAGE_SIZE = [224, 224]
resnet = ResNet50(
    include_top=False,
    weights='imagenet',
    input_shape=IMAGE_SIZE + [3] ## RGB Channels
)
```

```python
resnet.summary()
```

### Inital Steps

```python
train_path = 'Datasets/Train/'
valid_path = 'Datasets/Test/'
```

```python
folders = glob('Datasets/Train/*') # To get the number of classes/folders
```

```python
folders
```

```python
num_class = len(folders)
```

```python
num_class
```

```python
for layer in resnet.layers:
    layer.trainable = False #We dont want to retrain the weights
```

### Before going into the dense layers, the op needs to be flattened

```python
resnet.output
```

```python
x = Flatten()(resnet.output)
```

```python
x #flattened
```

```python
###
```

```python
prediction = Dense(num_class, activation='softmax')(x)
```

```python
prediction
```

```python
model = Model(inputs = resnet.input, outputs=prediction)
```

```python
model.summary()
```

```python
resnet.input
```

```python
prediction
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
##Data preprocessing
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, 
                                  shear_range=0.2,
                                  zoom_range=0.2,
                                  horizontal_flip=True
                                  )
test_datagen = train_datagen = ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory(directory=train_path, 
                                                target_size=(224,224),
                                                batch_size=32,
                                                class_mode='categorical'
                                                )
```

```python
test_set = test_datagen.flow_from_directory(directory=valid_path, 
                                                target_size=(224,224),
                                                batch_size=32,
                                                class_mode='categorical'
                                                )
```

```python
#fit the model
```

```python
r = model.fit(training_set,
                       validation_data=test_set,
                       epochs=50,
                       steps_per_epoch = len(training_set),
                       validation_steps=len(test_set)
                       )
```

```python
loss_df = pd.DataFrame(r.history)
```

```python
loss_df[['loss','val_loss']].plot()
```

```python
model.save('model_resnet50.h5')
```
